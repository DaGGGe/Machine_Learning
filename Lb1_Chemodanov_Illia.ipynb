```python
matrix = np.random.randint(1, 51, size=(9, 9))
divisible_by_3 = matrix % 3 == 0
percentage = (np.count_nonzero(divisible_by_3) / matrix.size) * 100

print("Матриця:")
print(matrix)
print(f"Відсоток чисел, які діляться на 3 без остачі: {percentage}%")

```

8.Створіть матрицю розміром 2x2, заповніть її випадковими цілими числами в діапазоні від 1 до 15 та знайдіть її обернену матрицю. 

```python
np.random.seed(0)  # Це для того, щоб результат був відтворюваним
matrix = np.random.randint(1, 16, size=(2, 2))

#Знаходимо обернену матрицю
inverse_matrix = np.linalg.inv(matrix)

print("Початкова матриця:")
print(matrix)
print("Обернена матриця:")
print(inverse_matrix)
```

```python
np.random.seed(0)  # Це для того, щоб результат був відтворюваним
matrix = np.random.randint(3, 16, size=(3, 3))

#Транспонтуемо матрицю
transposed_matrix = matrix.T

print("Початкова матриця:")
print(matrix)
print("Транспонована матриця:")
print(transposed_matrix)
```

14.Створіть матрицю розміром 9x9, заповніть її випадковими дійсними числами в діапазоні від -10 до 10 та знайдіть суму елементів на головній діагоналі. 

```python
# Створюемо матрицю розміром 9x9 та заповнюемо її випадковими дійсними числами від -10 до 10.
np.random.seed(0)  # Це для того, щоб результат був відтворюваним
matrix = np.random.uniform(-10, 10, size=(9, 9))

# Знаходимо суму елементів на головній діагоналі за допомогою функції np.trace.
sum_diagonal = np.trace(matrix)

# Виведимо матрицю та суму елементів на головній діагоналі.
print("Матриця:")
print(matrix)
print("Сума елементів на головній діагоналі:", sum_diagonal)
```

15.Створіть матрицю розміром 9x9, заповніть її випадковими дійсними числами в діапазоні від -0.5 до 0.5 та знайдіть вектор-стовпчик, що містить суму елементів кожного стовпця матриці. 

```python
# Створюемо матрицю розміром 9x9 та заповнюемо її випадковими дійсними числами від -0.5 до 0.5.
np.random.seed(0)  # Це для того, щоб результат був відтворюваним
matrix = np.random.uniform(-0.5, 0.5, size=(9, 9))

# Знаходимо суму елементів кожного стовпця матриці за допомогою функції np.sum і параметру axis=0.
column_sum = np.sum(matrix, axis=0)

# Перетворюемо результат вектор-стовпчик за допомогою .reshape.
column_sum_vector = column_sum.reshape(-1, 1)

# Виводимо матрицю та вектор-стовпчик сум.
print("Матриця:")
print(matrix)
print("Вектор-стовпчик сум:")
print(column_sum_vector)
```
