```python
import pandas as pd 

import numpy as np 

 

# Зчитуємо дані з вказаного URL 

url = "https://uk.wikipedia.org/wiki/%D0%A1%D0%B5%D1%80%D0%B5%D0%B4%D0%BD%D1%8F_%D0%B7%D0%B0%D1%80%D0%BE%D0%B1%D1%96%D1%82%D0%BD%D0%B0_%D0%BF%D0%BB%D0%B0%D1%82%D0%B0_%D0%B2_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D1%96" 

 

# Зчитуємо таблицю в DataFrame 

data = pd.read_html(url, header=0)[2]  # Обираємо другий елемент списку, де знаходиться таблиця 
```

```python
# Виводимо перші 5 рядків таблиці 

df_1 = data.head(5) 
print(df_1) 
```

```python
# Завдання 2: Визначте кількість рядків і стовпчиків за допомогою атрибуту shape.
shape = data.shape
print(f"Кількість рядків: {shape[0]}")
print(f"Кількість стовпчиків: {shape[1]}")

```

```python
# Завдання 3: Визначте загальну кількість років та кількість місяців у даних.
# Визначаємо загальну кількість років та місяців у даних
num_years = len(data['Рік']) - 1  # віднімання 1, оскільки перший стовпець - назва
num_months = num_years*12  
print("Загальна кількість років =", num_years)
print("Загальна кількість місяців =", num_months)
```

```python
# Завдання 4: Замініть пропущені значення "-" в таблиці на NaN.
# Також замінимо назви стовпців на зручні
data = data.rename(columns={
    "Рік": "Year",
    "В національній валюті": "Salary in UAH",
    "В доларах США": "Salary in USD",
    "Кількість кг хліба[20] на середню зарплату": "Bread per Salary",
    "Кількість БігМаків на середню зарплату": "BigMacs per Salary"
})
data = data.replace("—", np.nan)
data
```

```python
# Завдання 5: Визначте тип стовпців за допомогою dtypes.
column_types = data.dtypes

print(column_types)

```

```python
# 6. Видаліть передостанній і останній стовпці
data = data.drop(columns=["Bread per Salary", "BigMacs per Salary"])
print(data)
```

```python
#Видалимо * в 2022* і 2023*
data["Year"] = data["Year"].str.replace("*", "")
data["Year"] = data["Year"].astype(int)
data
```

```python
#Видалимо трикутники
data["Salary in UAH"] = data["Salary in UAH"].str.replace('▲', "").str.replace('▼', "")
data["Salary in USD"] = data["Salary in USD"].str.replace('▲', "").str.replace('▼', "")
data
```

```python
#Замінити тип нечислових даних на числові. 
column_types = data.dtypes
print(column_types)
data
```

```python
#Порахуйте частку пропусків в кожній колонці (за допомогою методів isnull , sum, len) 
missing_percentage = (data.isnull().sum() / len(data)) * 100
print("Частка пропусків в кожній колонці:")
print(missing_percentage)
```

```python
# Замінюємо пропущені значення в кожному стовпці на середнє значення цього стовпця
data_filled = data.fillna(data.mean())
data
```

```python
# Знаходимо рядок з максимальною зарплатою в Україні в гривнях
max_salary_row = data[data['Salary in UAH'] == data['Salary in UAH'].max()]

# Отримуємо рік та зарплату в Україні в гривнях для цього рядка
year_with_highest_salary = max_salary_row['Year'].values[0]
highest_salary_in_UAH = max_salary_row['Salary in UAH'].values[0]

print(f"Рік з найвищою зарплатою в Україні в гривнях: {year_with_highest_salary}")
print(f"Зарплата в Україні в гривнях: {highest_salary_in_UAH}")
```

```python
# Знаходимо рядок з мінімальною зарплатою в Україні в доларах США
min_salary_row = data[data['Salary in USD'] == data['Salary in USD'].min()]

# Отримуємо рік та зарплату в Україні в доларах США для цього рядка
year_with_lowest_salary = min_salary_row['Year'].values[0]
lowest_salary_in_USD = min_salary_row['Salary in USD'].values[0]

print(f"Рік з мінімальною зарплатою в Україні в доларах США: {year_with_lowest_salary}")
print(f"Зарплата в Україні в доларах США: {lowest_salary_in_USD}")
```

```python
# Обчислюємо середню зарплату в гривнях (UAH)
average_salary_UAH = data['Salary in UAH'].mean()

# Виводимо роки, коли зарплата в гривнях (UAH) була вище середньої
years_above_average_UAH = data[data['Salary in UAH'] > average_salary_UAH]['Year'].tolist()
print(f"Роки, коли зарплата в гривнях (UAH) була вище середньої: {years_above_average_UAH}")
```

```python
#Порівняйте  зарплату  в грн за останні два доступних роки.  
# Вибираємо дані за останні два доступних роки
latest_two_years_data = data.tail(2)

# Виводимо дані за останні два доступних роки
print(latest_two_years_data)

# Отримуємо зарплату в гривнях за останні два доступних роки
salary_UAH_last_year = latest_two_years_data.iloc[-2]['Salary in UAH']
salary_UAH_current_year = latest_two_years_data.iloc[-1]['Salary in UAH']

# Порівнюємо зарплату за останні два роки
if salary_UAH_current_year > salary_UAH_last_year:
    print(f"Зарплата в гривнях за поточний рік ({latest_two_years_data.iloc[-1]['Year']}) була вищою ніж за попередній рік ({latest_two_years_data.iloc[-2]['Year']}).")
elif salary_UAH_current_year < salary_UAH_last_year:
    print(f"Зарплата в гривнях за поточний рік ({latest_two_years_data.iloc[-1]['Year']}) була нижчою ніж за попередній рік ({latest_two_years_data.iloc[-2]['Year']}).")
else:
    print(f"Зарплата в гривнях за поточний рік ({latest_two_years_data.iloc[-1]['Year']}) була такою ж, як і за попередній рік ({latest_two_years_data.iloc[-2]['Year']}).")
```

```python
#Створіть графік, що відображає  зарплату в Україні з часом (за роками). 
import matplotlib.pyplot as plt

# Отримуємо дані про роки і зарплату в Україні
years = data['Year']
salary_UAH = data['Salary in UAH']

# Створюємо графік
plt.figure(figsize=(12, 6))
plt.plot(years, salary_UAH, marker='o', linestyle='-')

# Додаємо назви вісей та назву графіку
plt.title('Зарплата в Україні за роками')
plt.xlabel('Рік')
plt.ylabel('Зарплата в UAH')

# Додаємо сітку на графіку
plt.grid(True)

# Відображаємо графік
plt.show()
```

```python
#Видаліть останній рік з таблиці та виведіть оновлену таблицю. 
# Видалення останнього року (останнього рядка)
data = data.iloc[:-1]

# Виведення оновленої таблиці
print(data)
```

```python
#Побудуйте графік, що відображає середню зарплату в Україні в розрізі років без останнього року. 
import matplotlib.pyplot as plt

# Видалимо останній рік
data_without_last_year = data.iloc[:-1]

# Визначимо роки і середню зарплату
years = data_without_last_year['Year']
average_salary_UAH = data_without_last_year['Salary in UAH']

# Побудова графіка
plt.figure(figsize=(10, 6))
plt.plot(years, average_salary_UAH, marker='o', linestyle='-', color='b')
plt.title('Середня зарплата в Україні (грн) за роками')
plt.xlabel('Рік')
plt.ylabel('Середня зарплата (грн)')
plt.grid(True)
plt.xticks(years, rotation=45)
plt.tight_layout()

# Показати графік
plt.show()
```

```python
#Обчисліть зарплату за останні два роки і порівняйте її із зарплатою за останні 5 років. Вкажіть, чи є різниця, і в якому напрямку. 
# Отримаємо дані для останніх 5 років
last_5_years = data.tail(5)

# Отримаємо дані для останніх 2 років
last_2_years = last_5_years.tail(2)

# Обчислимо зарплату за останні 5 років та останні 2 роки
salary_last_5_years_UAH = last_5_years['Salary in UAH'].mean()
salary_last_2_years_UAH = last_2_years['Salary in UAH'].mean()

# Виведемо результати порівняння
if salary_last_2_years_UAH > salary_last_5_years_UAH:
    print(f"Середня зарплата за останні 2 роки ({salary_last_2_years_UAH:.2f} UAH) вища, ніж за останні 5 років ({salary_last_5_years_UAH:.2f} UAH).")
elif salary_last_2_years_UAH < salary_last_5_years_UAH:
    print(f"Середня зарплата за останні 2 роки ({salary_last_2_years_UAH:.2f} UAH) нижча, ніж за останні 5 років ({salary_last_5_years_UAH:.2f} UAH).")
else:
    print(f"Середня зарплата за останні 2 роки ({salary_last_2_years_UAH:.2f} UAH) дорівнює середній зарплаті за останні 5 років ({salary_last_5_years_UAH:.2f} UAH).")
```
